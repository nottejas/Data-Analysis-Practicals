```python
# Detecting and handling outliers using IQR
def removeoutliers(col):
    sorted(col)
    quant1,quant3=col.quantile([0.25,0.75])
    IQR=quant3-quant1
    LowerRange=quant1-1.5*IQR
    UpperRange=quant3 + 1.5*IQR
    return LowerRange,UpperRange
```

```python
lowscore,highscore=removeoutliers(df["Happiness Score"])
```

```python
df['Happiness Score'] = np.where(df['Happiness Score'] > highscore, highscore, df['Happiness Score'])
df['Happiness Score'] = np.where(df['Happiness Score'] < lowscore, lowscore, df['Happiness Score'])
```
