```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

titanic_data = pd.read_csv("Titanic (2).csv")
titanic_data
```

```python
titanic_data.nunique()
```

```python
titanic_data.drop(['PassengerId', 'Ticket'], axis=1, inplace=True)
```

```python
titanic_data
```

```python
titanic_data.info()
```

```python
titanic_data.describe()
```

```python
titanic_data.notnull()
```

```python
titanic_data.groupby(["Survived", "Sex"])["Survived"].count()
```

```python
sns.countplot(x="Survived", data=titanic_data)
plt.show()
```

```python
titanic_data[titanic_data['Sex'] == 'male'].Survived.groupby(titanic_data.Survived).count().plot(kind='pie',figsize=(6,6),explode=[0,0.05],autopct='%1.1f%%')
plt.axis('equal')
plt.legend(["Perished","Survived"])
plt.title("Male survival rate")
plt.show()
```

```python
pd.crosstab(titanic_data.Pclass, titanic_data.Survived, margins=True)
```

```python
titanic_data[titanic_data['Pclass'] == 1].Survived.groupby(titanic_data.Survived).count().plot(kind='pie',figsize=(6,6),explode=[0,0.05],autopct='%1.1f%%')
plt.axis('equal')
plt.legend(["Perished","Survived"])
plt.title("First class survival rate")
plt.show()
```

```python
pd.crosstab([titanic_data.Sex, titanic_data.Survived], titanic_data.Pclass, margins=True)
```

```python
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=titanic_data)
plt.show()
```

```python
sns.countplot(x='Survived', data=titanic_data, hue='Embarked')
plt.show()
```

```python
pd.crosstab([titanic_data.Sex, titanic_data.Survived], [titanic_data.SibSp, titanic_data.Pclass], margins=True)
```
