```python
corr = df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Corr heatmap of airlines")
plt.show()
```

```python
mean_performance = df.mean().sort_values(ascending=False)
print(mean_performance)

mean_performance.plot(kind='barh', color='skyblue', figsize=(10,6))
plt.title("Average Performance by Airline")
plt.xlabel("Average Value")
plt.show()
```
