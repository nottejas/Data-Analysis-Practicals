```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("setup comp")
```

```python
fifa_data = pd.read_csv("fifa.csv")
fifa_data

```

```python
plt.figure(figsize=(16,6))

sns.lineplot(data=fifa_data)
plt.show()
```

```python
fifa_data.info()
```

```python
fifa_data['Date'] = pd.to_datetime(fifa_data['Date'])
fifa_data.set_index("Date", inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(15, 7))

for country in fifa_data.columns:
    sns.lineplot(x=fifa_data.index, y=fifa_data[country], ax=ax, label = country)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 7))

for country in fifa_data.columns:
    sns.lineplot(x=fifa_data.index, y=fifa_data[country], ax=ax, label = country)

ax.invert_yaxis()
ax.set_title("FIFA Rankings", fontsize=14)
ax.set_xlabel("Year", fontsize=12)
ax.set_ylabel("Ranking", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Country")
plt.grid(True, linestyle="--",alpha=0.5)

plt.show()
```

```python
flight_delay = pd.read_csv("flight_delays.csv")
flight_delay.head()
```

```python
flight_delay.info()
```

```python
flight_delay.set_index('Month')
```

```python
plt.figure(figsize=(10, 6))

plt.title("Avg arrival delay for spirit airlines by month")
sns.barplot(x=flight_delay.index, y=flight_delay['NK'])
plt.show()
```

```python
plt.figure(figsize=(14, 7))

plt.title("Average arrival delay for each airline by month")

sns.heatmap(data=flight_delay, annot=True)

plt.xlabel("airline")
plt.show()
```

```python
# correaltion heatmap

flight_delay_corr = flight_delay.corr()
```

```python
plt.figure(figsize=(14,7))

plt.title("Correlation by month")

sns.heatmap(data=flight_delay_corr, annot=True, cmap="coolwarm")
plt.xlabel("Airline")
plt.show()
```

```python
iris_data = pd.read_csv("iris (1).csv")
iris_data
```

```python
sns.histplot(iris_data['Petal Length (cm)'])
plt.show()
```

```python
sns.histplot(iris_data['Petal Length (cm)'], kde=True)
plt.show()
```

```python
sns.kdeplot(data=iris_data['Petal Length (cm)'], shade=True)
plt.show()
```

```python
sns.jointplot(x=iris_data['Petal Length (cm)'],
             y=iris_data['Sepal Width (cm)'],
             kind="kde",
             cmap="coolwarm")
plt.show()
```

```python
sns.jointplot(x=iris_data['Petal Length (cm)'],
             y=iris_data['Sepal Width (cm)'],
             kind="kde").plot_joint(sns.scatterplot, color="red")
plt.show()
```

```python
sns.histplot(data=iris_data, x="Petal Length (cm)", hue="Species")
plt.title("Histogram of petal length")
plt.show()
```

```python
sns.kdeplot(data=iris_data, x="Petal Length (cm)", hue="Species", shade=True)
plt.title("Histogram of petal length")
plt.show()
```
