```python
wine = pd.read_excel('Wine_data.xlsx')
wine
```

```python
maax = wine['Hue'].max()
miin = wine['Hue'].min()

rangee = maax-miin

print("Range of hue is ",rangee)
```

```python
share = pd.read_csv("Share_data.csv")
share
```

```python
mean_o = share['Opening Price'].mean()
mean_c = share['Closing Price'].mean()

print('Mean of opening',mean_o)

print('Mean of closing',mean_c)
```

```python
median_o = share['Opening Price'].median()
median_c = share['Closing Price'].median()

print('Median of opening',median_o)

print('Median of closing',median_c)
```

```python
mode_o = share['Opening Price'].mode()
mode_c = share['Closing Price'].mode()

print('Mode of opening',mode_o)

print('Mode of closing',mode_c)
```

```python
sd_o = np.std(share['Opening Price'])
sd_c = np.std(share['Closing Price'])



print("Std of opening",sd_o)
print("Std of closing",sd_c)




```

```python
#rmse
x = share['Opening Price']
y = share['Closing Price']

best_fit = np.polyfit(x,y,1)

predicted_val = np.polyval(best_fit,x)

rmse= np.sqrt(np.mean((y-predicted_val)**2))

print("Best Fit is ",rmse)
```
