```python
import pandas as pd
import numpy as np

tumor_data = pd.read_csv("Tumor_data.csv")
tumor_data
```

```python
parameter_means = tumor_data.mean(numeric_only=True)

print("\nMean of all parameters:")
print(parameter_means)
```

```python
mean_radius = parameter_means['radius']

large_radius_tumors = tumor_data[tumor_data['radius'] > mean_radius]
print(large_radius_tumors)
```

```python
mean_radius=tumor_data['radius'].median()
median_perimeter=tumor_data['perimeter'].median()
special_tumors=tumor_data[(tumor_data['radius']<mean_radius)&(tumor_data['perimeter']>median_perimeter)]
special_tumors
```

```python
wine_data = pd.read_excel("Wine_data.xlsx")
wine_data
```

```python
# Calculate median values for alcohol and magnesium
alcohol_median = wine_data['Alcohol'].median()
magnesium_median = wine_data['Magnesium'].median()

# Filter wines where alcohol > median AND magnesium < median
high_alcohol_low_magnesium = wine_data[
    (wine_data['Alcohol'] > alcohol_median) & 
    (wine_data['Magnesium'] < magnesium_median)
]

print("\nWines with high alcohol but low magnesium:")
print(f"Alcohol median: {alcohol_median:.2f}, Magnesium median: {magnesium_median:.2f}")
print(high_alcohol_low_magnesium)
```

```python
def bubble_sort(students):
    n = len(students)
    for i in range(n):
        for j in range(0, n-i-1):
            # Compare marks
            if students[j][1] < students[j+1][1]:
                students[j], students[j+1] = students[j+1], students[j]
            # If marks are equal, compare names in DESCENDING order
            elif students[j][1] == students[j+1][1] and students[j][0] < students[j+1][0]:
                students[j], students[j+1] = students[j+1], students[j]
    return students

if __name__ == "__main__":
    n = int(input())
    students = []
    for _ in range(n):
        line = input().strip().split()
        name = line[0]
        marks = int(line[1])
        students.append((name, marks))

    sorted_students = bubble_sort(students)

    for student in sorted_students:
        print(student[0], student[1])
```
