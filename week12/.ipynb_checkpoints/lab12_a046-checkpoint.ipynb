```python
# Load Share Data
share_df = pd.read_csv('Share_data.csv')

# Mean, Median, Mode
mean_open = share_df['Opening Price'].mean()
median_open = share_df['Opening Price'].median()
mode_open = share_df['Opening Price'].mode()[0]

mean_close = share_df['Closing Price'].mean()
median_close = share_df['Closing Price'].median()
mode_close = share_df['Closing Price'].mode()[0]

print(f"Mean, Median, Mode of Opening Prices: {mean_open}, {median_open}, {mode_open}")
print(f"Mean, Median, Mode of Closing Prices: {mean_close}, {median_close}, {mode_close}")

# Standard Deviation
std_open = share_df['Opening Price'].std()
std_close = share_df['Closing Price'].std()
print(f"Standard deviation of Opening Prices: {std_open}")
print(f"Standard deviation of Closing Prices: {std_close}")

# RMSE with respect to best fit line (assumed as mean value)
rmse_open = np.sqrt(((share_df['Opening Price'] - mean_open) ** 2).mean())
rmse_close = np.sqrt(((share_df['Closing Price'] - mean_close) ** 2).mean())
print(f"RMSE for Opening Prices: {rmse_open}")
print(f"RMSE for Closing Prices: {rmse_close}")
```

```python
x=share_df['Opening Price']
y=share_df['Closing Price']
best_fit=np.polyfit(x,y,1)
pridicated_values=np.polyval(best_fit,x)
rmse=np.sqrt(np.mean((y-pridicated_values)**2))
rmse
```

```python
#Task 2) #Maximum absolute scaling
numerical_cols = ['age', 'salary', 'performance score']
max_abs_df = emp_df.copy()

for col in numerical_cols:
    max_abs = max_abs_df[col].abs().max()
    max_abs_df[col] = max_abs_df[col] / max_abs

print(max_abs_df[['id'] + numerical_cols])
```

```python
#Min-max feature scaling
min_max_df = emp_df.copy()

for col in numerical_cols:
    min_val = min_max_df[col].min()
    max_val = min_max_df[col].max()
    if min_val != max_val:
        min_max_df[col] = (min_max_df[col] - min_val) / (max_val - min_val)

print(min_max_df[['id'] + numerical_cols])
```

```python
#z-score

z_score_df = emp_df.copy()

for col in numerical_cols:
    mean = z_score_df[col].mean()
    std = z_score_df[col].std()
    z_score_df[col] = (z_score_df[col] - mean) / std

print(z_score_df[['id'] + numerical_cols])
```

```python
plt.imshow(img_array)
plt.show()
```

```python
block_size = (2,2)
```

```python
height -= width % block_size[0]
width -= width % block_size[1]
```

```python
blocks = [
    img_array[i:i+block_size[0], j:j+block_size[1]]
    for i in range(0, height, block_size[0])
    for j in range(0, width, block_size[1])
]
blocks
```

```python
blocks_array=np.array([block.flatten() for block in blocks])

print("Shape of 2D blocks array",blocks_array.shape)

print(blocks_array[:5])
 
```

```python
#further split group1 and group 2 based on mean of the 2nd column
mean_col1=np.mean(blocks_array[:,[0,2]],axis=1)
split_index1=np.mean(mean_col1)
group1 = blocks_array[mean_col1 <= split_index1]
group2 = blocks_array[mean_col1 > split_index1]
print(group1)
```

```python
mean_col2_group1=np.mean(group1[:,[1,3]],axis=1)
mean_col2_group2=np.mean(group2[:,[1,3]],axis=1)
 
split_index2_group1=np.mean(mean_col2_group1)
split_index2_group2=np.mean(mean_col2_group2)
 
group1a=group1[mean_col2_group1<=split_index2_group1]
 
group1b=group1[mean_col2_group1>split_index2_group1]
 
group2a=group2[mean_col2_group2<=split_index2_group2]
 
group2b=group2[mean_col2_group2>split_index2_group2]
print(group1a)
print(group1b)
print(group2a)
print(group2b)
```
