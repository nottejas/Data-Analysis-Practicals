```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
tumor_data = pd.read_csv("tumor_data.csv")
tumor_np = tumor_data.to_numpy()
tumor_np
```

```python
# Find mean of all parameters
mean_values = np.mean(tumor_np, axis=0)
print("Mean of all parameters:", mean_values)
```

```python
mean_radius = mean_values[0]  
large_tumors = tumor_np[tumor_np[:, 0] > mean_radius]
print("Tumors greater than mean radius:", large_tumors)
```

```python
# Load Share Data
share_df = pd.read_csv('Share_data.csv')

# Mean, Median, Mode
mean_open = share_df['Opening Price'].mean()
median_open = share_df['Opening Price'].median()
mode_open = share_df['Opening Price'].mode()[0]

mean_close = share_df['Closing Price'].mean()
median_close = share_df['Closing Price'].median()
mode_close = share_df['Closing Price'].mode()[0]

print(f"Mean, Median, Mode of Opening Prices: {mean_open}, {median_open}, {mode_open}")
print(f"Mean, Median, Mode of Closing Prices: {mean_close}, {median_close}, {mode_close}")

# Standard Deviation
std_open = share_df['Opening Price'].std()
std_close = share_df['Closing Price'].std()
print(f"Standard deviation of Opening Prices: {std_open}")
print(f"Standard deviation of Closing Prices: {std_close}")

# RMSE with respect to best fit line (assumed as mean value)
rmse_open = np.sqrt(((share_df['Opening Price'] - mean_open) ** 2).mean())
rmse_close = np.sqrt(((share_df['Closing Price'] - mean_close) ** 2).mean())
print(f"RMSE for Opening Prices: {rmse_open}")
print(f"RMSE for Closing Prices: {rmse_close}")
```

```python
x=share_df['Opening Price']
y=share_df['Closing Price']
best_fit=np.polyfit(x,y,1)
pridicated_values=np.polyval(best_fit,x)
rmse=np.sqrt(np.mean((y-pridicated_values)**2))
rmse
```
