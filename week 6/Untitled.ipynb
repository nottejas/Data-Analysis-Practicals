```python
# 3
cricket_play = set(cricket_data['ID']) 
volley_play = set(volley_data['ID']) 
hockey_play= set(hockey_data['ID'])

only_cricket = cricket_play - (volley_play | hockey_play)
only_volley = volley_play - (cricket_play | hockey_play)
only_hockey = hockey_play - (cricket_play | volley_play)

only_one_event = only_cricket | only_volley | only_hockey

final_one_event = master_data[master_data['ID'].isin(only_one_event)][['ID', 'NAME']]
final_one_event
```
