```python
car_data = car_data.drop(['S.No.'], axis=1)
car_data
```

```python
from datetime import date
date.today().year
car_data['Car Age']=date.today().year - car_data['Year']
car_data
```

```python
car_data['Brand'] = car_data.Name.str.split().str.get(0)
car_data['Model'] = car_data.Name.str.split().str.get(1) + ' ' + car_data.Name.str.split().str.get(2)
car_data
```

```python
car_data['Brand'].unique()
```

```python
searchfor = ['Isuzu', 'ISUZU', 'Mini', 'Land']
car_data[car_data.Brand.str.contains('|'.join(searchfor))].head(20)
```

```python
car_data['Brand'].replace({"ISUZU": "Isuzu", "Mini": "Minicooper", "Land": "Land Rover"}, inplace=True)
car_data["Brand"].head(10)
car_data.Brand.unique()
```

```python
car_data['Engine'] = car_data['Engine'].str.extract('(\d+)').astype(float)
car_data
```

```python
cat_cols = car_data.select_dtypes(include=['object']).columns
num_cols = car_data.select_dtypes(include=np.number).columns.tolist()
cat_cols
num_cols
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for col in num_cols:
    print(col)
    print("Skew: ", round(car_data[col].skew(), 2))

    plt.figure(figsize=(15, 4))

    plt.subplot(1, 2, 1)
    car_data[col].hist(grid=False)
    plt.ylabel('Count')

    plt.subplot(1, 2, 2)  
    sns.boxplot(x=car_data[col])  

    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(3, 2, figsize=(18, 18))
fig.suptitle("Bar plot for categorical variables")

sns.countplot(ax=axes[0, 0], x='Fuel_Type', data=car_data, color="blue",
              order=car_data['Fuel_Type'].value_counts().index)

sns.countplot(ax=axes[0, 1], x='Transmission', data=car_data, color="blue",
              order=car_data['Transmission'].value_counts().index)

sns.countplot(ax=axes[1, 0], x='Owner_Type', data=car_data, color="blue",
              order=car_data['Owner_Type'].value_counts().index)

sns.countplot(ax=axes[1, 1], x='Location', data=car_data, color="blue",
              order=car_data['Location'].value_counts().index)

sns.countplot(ax=axes[2, 0], x='Brand', data=car_data, color="blue",
              order=car_data['Brand'].value_counts().index)

sns.countplot(ax=axes[2, 1], x='Model', data=car_data, color="blue",
              order=car_data['Model'].value_counts().index)
axes[2, 1].axis("off")

plt.tight_layout(rect=[0, 0, 1, 0.96])  
plt.show()
```

```python
plt.figure(figsize=(13,17))
sns.pairplot(data=car_data.drop(['Kilometers_Driven', 'Price'],axis=1))
plt.show()
```

```python


fig,axarr = plt.subplots(4,2,figsize=(12,18))
car_data.groupby('Location')['Price'].mean().sort_values(ascending=False).plot.bar(ax=axarr[0][0],fontsize=5)
axarr[0][0].set_title("Location Vs Price",fontsize=18)
 
car_data.groupby('Transmission')['Price'].mean().sort_values(ascending=False).plot.bar(ax=axarr[0][1],fontsize=5)
axarr[0][1].set_title("Transmission Vs Price",fontsize=18)
 
car_data.groupby('Fuel_Type')['Price'].mean().sort_values(ascending=False).plot.bar(ax=axarr[1][0],fontsize=5)
axarr[1][0].set_title("Fuel_Type Vs Price",fontsize=18)
 
car_data.groupby('Owner_Type')['Price'].mean().sort_values(ascending=False).plot.bar(ax=axarr[1][1],fontsize=5)
axarr[1][1].set_title("Owner_Type Vs Price",fontsize=18)
 
car_data.groupby('Brand')['Price'].mean().sort_values(ascending=False).plot.bar(ax=axarr[2][0],fontsize=5)
axarr[2][0].set_title("Brand Vs Price",fontsize=18)
 
car_data.groupby('Model')['Price'].mean().sort_values(ascending=False).plot.bar(ax=axarr[2][1],fontsize=5)
axarr[2][1].set_title("Model Vs Price",fontsize=18)
 
car_data.groupby('Seats')['Price'].mean().sort_values(ascending=False).plot.bar(ax=axarr[3][0],fontsize=5)
axarr[3][0].set_title("Seats Vs Price",fontsize=18)
 
car_data.groupby('Car Age')['Price'].mean().sort_values(ascending=False).plot.bar(ax=axarr[3][1],fontsize=5)
axarr[3][1].set_title("Car Age Vs Price",fontsize=18)
 
plt.subplots_adjust(hspace=1.0)
plt.subplots_adjust(wspace=.5)
sns.despine()

```

```python
plt.figure(figsize=(18, 10))

corr_matrix = car_data[['Kilometers_Driven', 'Price']].corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

plt.title("Correlation Heatmap: Kilometers Driven vs Price")
plt.show()
```
