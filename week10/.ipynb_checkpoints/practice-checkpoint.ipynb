```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
data = pd.read_csv('used_cars_data.csv')
data.head()
```

```python
data.set_index('S.No.', inplace=True)
data
```

```python
data.info()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
data['Mileage']=data['Mileage'].str.replace("kmpl","",regex=False)
data['Mileage']=data['Mileage'].str.replace("km/kg","",regex=False)
data.head()
```

```python
data.info()
```

```python
data['Mileage']=data['Mileage'].astype(float)
data.info()
```

```python
data.info()
```

```python
data['Engine']=data['Engine'].str.replace("CC","",regex=False)
data
```

```python
# data['Engine'].fillna(data['Engine'].mean,inplace=True)
data['Engine']=data['Engine'].astype(float)
data
```

```python
data['Engine'].head()
```

```python
data['Power']=data['Power'].str.replace("bhp","", regex=False)
data
```

```python
data.isnull().sum()/len(data)*100
```

```python
data.drop("New_Price",inplace=True,axis=1)
data
```

```python
data['Price'].fillna(data["Price"].mean(),inplace=True)
data
```

```python
import datetime as dt
```

```python
data['Age']=dt.date.today().year-data['Year']
data
```

```python
grp_loc = data.groupby("Location")['Price'].mean()
grp_loc
```

```python
sb.barplot(grp_loc)
```

```python
data.boxplot(column='Mileage')
plt.show()
```

```python
def outlier(col):
    sorted(col)
    quant1,quant3=col.quantile([0.25,0.75])
    IQR=quant3-quant1
    lowerRange=quant1-1.5*IQR
    upperRange=quant3+1.5*IQR
    return upperRange,lowerRange
```

```python
lowerscore,highscore=outlier(data['Mileage'])
data['Mileage']=np.where(data['Mileage']>highscore,highscore,data['Mileage'])
data['Mileage']=np.where(data['Mileage']<lowerscore,lowerscore,data['Mileage'])
```

```python
data.boxplot(column="Mileage")
plt.show()
```
