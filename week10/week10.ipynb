```python
def outlier(col):
    sorted(col)
    q1, q3 = col.quantile([0.25, 0.75])
    IQR = q3-q1
    lowerRange=q1-1.5*IQR
    upperRange=q3+1.5*IQR
    return upperRange, lowerRange
```

```python
lowsc, highsc = outlier(df['Mileage'])
df['Mileage']=np.where(df['Mileage']>highsc, highsc, df['Mileage'])
df['Mileage']=np.where(df['Mileage']>lowsc, lowsc, df['Mileage'])
```

```python
df.boxplot('Mileage')
plt.show()
```
