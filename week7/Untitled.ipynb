```python
string = input("Enter a string: ").lower()
vowels = 'aeiou'
vowel_freq = {}

for char in string:
    if char in vowels:
        vowel_freq[char] = vowel_freq.get(char, 0) + 1

total_vowels = sum(vowel_freq.values())

print("Vowel Frequency:", vowel_freq)
for vowel, freq in vowel_freq.items():
    percentage = (freq / total_vowels) * 100
    print(f"{vowel}: {percentage:.2f}%")
```
