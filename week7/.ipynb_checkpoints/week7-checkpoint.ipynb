```python
import pandas as pd

df = pd.read_excel("Master_data_prac7.xlsx")
df
```

```python
#1
max_score = df.groupby('dept')['sport score'].max().reset_index()
max_score
```

```python
#2
age_mini = df.loc[df.groupby('city')['age'].idxmin(), ['city', 'name', 'age']]
age_mini
```

```python
#3 
min_max_age = df.groupby(['city', 'dept'])['age'].agg(['min', 'max'])
min_max_age
```

```python
#4
average_score = df.groupby(['city', 'dept'])['sport score'].mean()
average_score
```

```python
# 5
crick_df = df[df['sport'] == 'cricket']
young = crick_df.loc[crick_df['sport score'].idxmax()]
young_deg = young['designation']
young_deg
```

```python
#6
elder = df.loc[df['sport score'].idxmax()]
elder_loc = elder[['city', 'dept']]
elder_loc
```

```python
sport = df.groupby(['city', 'dept'])['sport score'].mean().reset_index()
score = sport.loc[sport.groupby('city')['sport score'].idxmax()]
print(score[['city', 'dept', 'sport score']])
```
