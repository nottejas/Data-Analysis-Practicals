```python
import pandas as pd

with open("faculty_data.txt", "r") as file:
    lines = file.readlines()

name, position, dept, degree, exp, address = [], [], [], [], [], []

for i, line in enumerate(lines):
    clean_line = line.strip()
    if i % 6 == 0:
        name.append(clean_line)
    elif i % 6 == 1:
        position.append(clean_line)
    elif i % 6 == 2:
        dept.append(clean_line)
    elif i % 6 == 3:
        degree.append(clean_line)
    elif i % 6 == 4:
        exp.append(clean_line)
    elif i % 6 == 5:
        address.append(clean_line)

df_faculty = pd.DataFrame({
    "Name": name,
    "Position": position,
    "Department": dept,
    "Degree": degree,
    "Experience": exp,
    "Address": address,
})

print(df_faculty)
```

```python
df_pul = pd.read_csv("Pulse_rate_data.csv")
df_pul.isnull().sum().sum()
```

```python
df_formatted = df_pul.copy()                 
df_formatted
```

```python
df_formatted['Date'].fillna('No data', inplace=True)
df_formatted['Pulse'].fillna(df_formatted['Pulse'].mean(),inplace=True)
df_formatted['Maxpulse'].fillna(df_formatted['Maxpulse'].mean(),inplace=True)
df_formatted['Calories'].fillna(df_formatted['Calories'].mean(),inplace=True)
df_formatted  
```
